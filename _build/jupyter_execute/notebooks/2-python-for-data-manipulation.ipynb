While Python itself can manipulate data, specialised libraries make the process efficient. ### Data manipulation

Data manipulation refers to transforming data into a format that is easier to read, interpret, and use. This may involve:

- Renaming variables
- Filtering and sorting observations
- Creating new variables
- Handling missing values
- Merging or reshaping datasets

The `pandas` library provides efficient and intuitive functions for all these tasks.

### Data analysis

Data analysis is the process of examining data to uncover patterns, trends, and insights that can inform decision-making. In Python, this is typically achieved by combining `pandas` (for data handling) with visualisation tools and statistical methods.

### Ordinary least squares regression with statsmodels

This example examines a regression of population against industry. With `statsmodels`, you first create a model before fitting it to the data.

````python
import statsmodels.formula.api as smf
````

````python
pollute_model = smf.ols('Population ~ Industry', data=pollute_data)
````

We can once again use `help` to gain insight:

````python
help(pollute_model)
````

First, notice how the model is specified using the string 'Population ~ Industry'. The formula is used to generate components such as the design matrix, but the model is not fitted at this stage. To explore what the model contains, you can use Python’s `dir` function.

````python
dir(pollute_model)
````

One of the methods, `fit`, fits the model to the data.

````python
pollute_fit = pollute_model.fit()
````

#### Plotting results

Like `pandas`, `statsmodels` also has some built-in visualisation functionalities.

````python
import statsmodels.api as sm
````

````python
sm.graphics.plot_regress_exog(pollute_fit, "Industry").tight_layout(pad=1.0)
````

````python
sm.qqplot(pollute_fit.resid);
````

````python
sm.graphics.plot_fit(pollute_fit,"Industry");
````

### Hypothesis testing with `scipy`

The `statsmodels` library is very useful for building data models, but for hypothesis testing, the `scipy` library is generally preferred.

#### Kolmogorov-Smirnov tests

The wings data set contains count data for the size of insects wings from two different locations.

````python
wings_data
````

This example tests the hypothesis that the distribution of wing sizes from the two locations are the same.

````python
import scipy
````

````python
res = scipy.stats.kstest(wings_data['LocationA'],wings_data['LocationB'])
print(res)
````
