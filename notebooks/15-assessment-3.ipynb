## Part 3.3

If $n$ points are randomly placed in a region with area $A$, the expected distance from each point to its nearest neighbour is approximately:

$$
\bar{r}_\text{expected} = \frac{1}{2} \sqrt{\frac{A}{n}}
$$

[Clark and Evans (1954)](https://esajournals.onlinelibrary.wiley.com/doi/10.2307/1931034) derived this formula in spatial statistics. 

- Compute this theoretical value for your $n$ and $A$.
- Compare with your simulated average nearest-neighbour (NN) distance.
- Compute the Clark & Evans $R$ statistic:

$$
R = \frac{\text{observed mean NN distance}}{\text{expected NN distance}}
$$

Comment on the value of $R$ and what does it tell us about the randomness of the data.
